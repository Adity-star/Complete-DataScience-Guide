```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch
import re
```

```python
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
model = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone')
```

```python
def finbert_sentiment(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
    outputs = model(**inputs)
    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=-1).item()

    if predicted_class == 0:
        return "Negative"
    elif predicted_class == 1:
        return "Neutral"
    else:
        return "Positive"
```

```python
def preprocess_text(text):
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = text.strip()  # Remove leading and trailing spaces
    return text
```

```python
news_article = """
    The U.S. stock market experienced a significant rally today following strong earnings reports from tech giants.
    Investors are optimistic about the future as companies continue to show robust growth despite global economic challenges.
"""

# Preprocess the text
clean_article = preprocess_text(news_article)

# Get the sentiment analysis result using FinBERT
sentiment = finbert_sentiment(clean_article)
print(f"Sentiment of the article: {sentiment}")
```

```python
news_articles = [
    "Tech stocks surged after strong earnings reports, while traditional industries lagged behind.",
    "The Federal Reserve's interest rate hike led to a drop in stock prices, causing widespread market concerns.",
    "Investors are hopeful after a strong quarterly earnings season, with tech stocks leading the charge."
]

# Analyze sentiment for all articles
sentiments = [finbert_sentiment(preprocess_text(article)) for article in news_articles]

# Count occurrences of each sentiment
positive = sentiments.count("Positive")
neutral = sentiments.count("Neutral")
negative = sentiments.count("Negative")

# Print the aggregated results
print(f"Positive: {positive}, Neutral: {neutral}, Negative: {negative}")
```
