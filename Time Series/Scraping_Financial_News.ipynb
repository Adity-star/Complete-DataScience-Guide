### For extracting from different sections according to your use like
* sentiment analysis
* stock price prediction
* Trend Monitoring

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = 'https://www.moneycontrol.com/news/'

sections = [
    'market-news',
    'corporate-news',
    'economic-news',
    'general-news',
]

all_news_data = {section: [] for section in sections}

def fetch_news_from_section(section_name, section_path, page_number):

    url = f"{base_url}{section_path}/page/{page_number}/"
    print(f"Fetching news from {section_name} - Page {page_number}...")

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    news_section = soup.find('section', class_='startup-news section')

    if news_section:
        for news_item in news_section.find_all('h3', class_='related_des'):
            headline = news_item.text.strip()
            all_news_data[section_name].append({
                'Date': '2023-10-04',
                'Section': section_name,
                'Headline': headline
            })

for section, section_path in zip(sections, sections):

    for page_number in range(1, 3):
        fetch_news_from_section(section, section_path, page_number)

# Store the data in csv files for further use
for section, news_list in all_news_data.items():
    df = pd.DataFrame(news_list)
```

```python
 # Save DataFrame to CSV
file_name = f"{section.replace('-', '_')}_headlines.csv"
df.to_csv(file_name, index=False, encoding='utf-8')
print(f"Data for {section} has been written to '{file_name}'.")
```
