```python
#3D line plots
x = [0,1,5,25]
y = [0,10,15,0]
z = [0,13,20,23]

fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='red')
plt.show()
```

```python
# 3D surface plots
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
bat = pd.read_csv(r"C:\Users\Administrator\Datasets\batsman_season_record.csv")
bat
```

```python
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Plot for 2015
axs[0].bar(range(len(bat['2015'])), bat['2015'])
axs[0].set_title('Batsman Data for 2015')
axs[0].set_ylabel('Batsman')
axs[0].set_xlabel('Index')

# Plot for 2016
axs[1].bar(range(len(bat['2016'])), bat['2016'])
axs[1].set_title('Batsman Data for 2016')
axs[1].set_ylabel('Batsman')
axs[1].set_xlabel('Index')

# Plot for 2017
axs[2].bar(range(len(bat['2017'])), bat['2017'])
axs[2].set_title('Batsman Data for 2017')
axs[2].set_ylabel('Batsman')
axs[2].set_xlabel('Index')

# Adjust layout to avoid overlapping labels
plt.tight_layout()

plt.show()
```

```python
# Plotting the data as a bar chart
bat.plot(kind='bar', figsize=(8, 6))

# Customizing the plot
plt.title('Batsman Data for 2015, 2016, and 2017')
plt.ylabel('Batsman')
plt.xlabel('Index')

plt.show()
```

```python
# Plotting the stacked bar chart
bat.plot(kind='bar', stacked=True, figsize=(8, 6))

# Customizing the plot
plt.title('Stacked Batsman Data for 2015, 2016, and 2017')
plt.ylabel('Batsman')
plt.xlabel('Index')

# Displaying the plot
plt.show()
```

```python
# Plotting each year as a separate pie chart
bat[['2015', '2016', '2017']].plot(kind='pie', subplots=True, figsize=(15, 8), legend=False, autopct='%1.1f%%')

# Customizing the plot
plt.suptitle('Pie Charts for Batsman Data (2015, 2016, and 2017)', fontsize=16)

# Displaying the plot
plt.show()
```
