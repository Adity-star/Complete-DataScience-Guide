### Processing the data

- `AutoTokenizer`: Used to load pre-trained tokenizers for transforming text into token representations.
- `AutoModelForSequenceClassification`: Used to load pre-trained models for sequence classification tasks.
- `AdamW`: An optimizer used for training models, specifically implementing the Adam optimization algorithm with weight decay.

2. **Model Initialization:**
   - `checkpoint = 'distilbert-base-uncased'`: Specifies the pre-trained model to be used, which is `distilbert-base-uncased`, a smaller version of BERT that doesn't differentiate between uppercase and lowercase letters.

- **Tokenization**: The process of converting raw text into tokens that a model can understand.
- **Pre-trained Model**: A model that has already been trained on large datasets and can be fine-tuned or directly used for specific tasks.
- **Sequence Classification**: A task where a model is used to classify sequences of text into categories (e.g., sentiment classification, spam detection).



```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification,AdamW

checkpoint = 'distilbert-base-uncased'
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

# Loading the model
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
```

preparing your input sequence to be fed into the transformer model by tokenizing it, ensuring consistent lengths through padding or truncation, and converting it into a format (PyTorch tensors) that can be used in the model.

- **`sequence`**: The input text (could be a single string or a list of strings) to be tokenized.
- **`padding=True`**: Adds padding to the input sequence(s) to make sure they are of equal length. Padding is added to sequences shorter than the longest one in the batch.
- **`truncation=True`**: Ensures that sequences longer than the model's maximum token length are truncated to fit within the model's limits.
- **`return_tensors='pt'`**: Returns the tokenized output as PyTorch tensors (`'pt'` stands for PyTorch). This makes the data ready for input into a PyTorch-based model.


```python
sequence = ["I've been waiting for this oppurtinaty my whole life.",
            "I hate this so much!"]

batch = tokenizer(sequence, padding = True, truncation = True, return_tensors = 'pt')
```

```python
batch
```

```python
batch['labels'] = torch.tensor([1,1])
batch
```

```python
optimizer = AdamW(model.parameters())

loss = model(**batch).loss
loss.backward()
optimizer.step()
```

#### Train our dataset and finetune it

- **TrainingArguments**: A class that holds hyperparameters and configurations for training a model using the `Trainer` API in Hugging Face's `transformers` library.
- **`"test-trainer"`**: This is the directory name where the model's training-related files (logs, checkpoints, etc.) will be stored. It is useful for tracking and resuming training, and for saving progress.


```python
from transformers import TrainingArguments

training_args = TrainingArguments("test-trainer")
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(checkpoint,num_labels = 2).to(device)
```


- **`model`**: The model to be trained or fine-tuned (e.g., `distilbert-base-uncased`).
- **`training_args`**: The `TrainingArguments` object that contains all the hyperparameters and configuration for the training process (batch size, number of epochs, learning rate, etc.).
- **`train_dataset = tokenized_datasets["train"]`**: The tokenized training dataset.
- **`eval_dataset = tokenized_datasets["validation"]`**: The tokenized validation dataset, which is used to evaluate the model's performance during training.
- **`data_collator = data_collator`**: A data collator that ensures sequences in the batch are padded to the same length.

```python
from transformers import Trainer

trainer = Trainer(
    model,
    training_args,
    train_dataset = tokenized_datasets["train"],
    eval_dataset = tokenized_datasets["validation"],
    data_collator = data_collator,
    tokenizer = tokenizer,
)
```

```python
trainer.train()
```

```python
predictions = trainer.predict(tokenized_datasets["validation"])
print(predictions.predictions.shape, predictions.label_ids.shape)
```

```python
import numpy as np

preds = np.argmax(predictions.predictions, axis=-1)
```

```python
def compute_metrics(eval_preds):
    metric = evaluate.load("glue", "mrpc")
    logits, labels = eval_preds
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
training_args = TrainingArguments("test-trainer", evaluation_strategy="epoch")
model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2).to(device)

trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```
